```python
import matplotlib.pyplot as plt

import random
import numpy as np
```

# 摺線圖

```python
x=[random.randint(0, 100) for _ in range(10)]

plt.plot(x)
plt.show()
```

# 散佈圖

```python
x=np.array([random.randint(0, 100) for _ in range(100)])
y= x*[random.uniform(0.8, 1.2) for _ in range(100)]

plt.scatter(x, y)
plt.show()
```

```python
x=np.array([random.randint(0, 100) for _ in range(100)])
y=x*[random.uniform(0.8, 1.2) for _ in range(100)]
s=y**1.5

plt.scatter(x, y, s)
plt.show()
```

# 長條圖

```python
x= [random.uniform(0, 100) for _ in range(100)]
y= list(range(100))

plt.bar(x, y)
plt.show()
```

# 盒狀圖

```python
x=np.random.normal(size = 100000)

plt.boxplot(x)
plt.show()
```

# 直方圖

```python
x=np.random.normal(size = 100000)

plt.hist(x)
plt.show()
```

# 設置 xlim, ylim

```python
x=[random.randint(50, 100) for _ in range(10)]

plt.plot(x)
plt.xlim(0, 15)
plt.ylim(0, 100)
plt.show()
```

# mark

https://matplotlib.org/api/markers_api.html

```python
x=[random.randint(50, 100) for _ in range(10)]

plt.plot(x, '*', markevery =([1,4]), markersize=30)

plt.plot(x)
plt.show()
```
