```python
import cv2
import matplotlib.pyplot as plt

import numpy as np
```

# read image as numpy array

```python
# numpy array with BGR format
img = cv2.imread('cat.jpeg')

# numpy array with RGB format
img = plt.imread('cat.jpeg')

print(img.shape)
```

# save image

```python
img = cv2.imread('cat.jpeg')
cv2.imwrite('cat.png', img)
```

# print image

```python
plt.imshow(img)
plt.show()
```

# image resize

```python
tmp = cv2.resize(img, (256, 256))

plt.imshow(tmp)
plt.show()
```

# rotate 90

```python
tmp = np.rot90(img, 1)

plt.imshow(tmp)
plt.show()
```

# clip

```python
tmp = img[:, :1500]

plt.imshow(tmp)
plt.show()
```

# to grayscale

```python
tmp = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(tmp, cmap='gray')
plt.show()
```

```python
tmp = cv2.imread('cat.jpeg', 0)

plt.imshow(tmp, cmap='gray')
plt.show()
```

# thresholding 

```python
tmp = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

ret,tmp = cv2.threshold(tmp,127,255,cv2.THRESH_BINARY)

plt.imshow(tmp, cmap='gray')
plt.show()
```

```python
tmp = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

ret,tmp = cv2.threshold(tmp, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(tmp, cmap='gray')
plt.show()
```

# add noise

```python
noise = [np.random.randint(0, i - 1, int(100000)) for i in img.shape[:2]]
tmp = img[:]
tmp[noise] = 255

plt.imshow(tmp)
plt.show()
```

# blur

```python
tmp = cv2.GaussianBlur(img,(51, 51), 0)

plt.imshow(tmp)
plt.show()
```

```python
tmp = cv2.medianBlur(img, 21)

plt.imshow(tmp)
plt.show()
```

```python
# add noise
noise = [np.random.randint(0, i - 1, int(100000)) for i in img.shape[:2]]
tmp = img[:]
tmp[noise] = 255

# median blur
tmp = cv2.medianBlur(img, 5)

plt.imshow(tmp)
plt.show()
```

# edge detection

```python
tmp = cv2.imread('cat2.jpg', 0)
edges = cv2.Canny(tmp,100,200)

plt.imshow(tmp, cmap='gray')
plt.show()
plt.imshow(edges, cmap='gray')
plt.show()
```
